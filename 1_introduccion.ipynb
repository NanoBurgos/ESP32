# **¿Que es el ESP32?**

Es un chip de la familia SoC, este chip incluye WiFi, Bluetooh y otras caracteristicas que veremos mas adelante.

El ESP32 es el sucesor del ESP8266

### **El Chip ESP32 se puede usar de dos maneras diferentes:**

- De forma independiente para agregarlo a placas personalizadas



- O montado en un kit de desarrollo listo para usar



Existen muchos kits de desarrollo independientes que incluyen el chip ESP32 en sus placas



## **Uno de los mas conocidos es el ESP32- DEVKIT V1** 

### ENTRADAS/SALIDAS

- Su voltaje de alimentacion es de 5 voltios cuando usamos el pin micro usb o el pin VIN

- Pero todos sus pines de entradas o salidas funcionan a 3.3 voltios

Esto se debe de tener en cuenta si sera conectado a otras placas o sensores, en esos casos se deben usar conversores de nivel logico 

### CEREBRO PRINCIPAL

- El cerebro principal de esta placa es un procesador de dos nucleos de 32 bits, en concreto el procesador **Xtensa Dual-Core 32-bit LX6 con 600 DMIPS**

- Tambien dispone de un co-procesador de ultra bajo consumo que es utilizado para realizar ciertas operaciones cuando el dispositivo se encuentra en modo DEEP SLEEP o tambien conocido como sueño profundo, en este modo el chip reduce su concumo energetico a unos 10 micro amperios, esto lo hace ideal para el uso de baterias debido a su bajo consumo

### FRECUENCIA DE OPERACION

- La velocidad de su reloj interno es de 240 mega hercios, es decir, ejecuta instrucciones 15 veces mas rapido que una placa arduino uno

### OPERACIONES

- Nos permite realizar operaciones con numeros reales de manera muy eficiente, asi como realizar multiplicaciones de numeros muy grandes de forma instantanea

### DIFERENCIAS CON OTROS MICRO CONTROLADORES

A difrencia de la mayoria de microcontroladores basados en las placas de arduino que solo disponen de 3 tipos de memorias (RAM, FLASH Y EEPROM)

**¿Qué es EEPROM?**
- EEPROM Significa memoria de solo lectura programable y borrable eléctricamente. Conserva los datos incluso cuando se corta la energía. A diferencia de las memorias convencionales. ROM, puedes borrar y reescribir EEPROM Utiliza electricidad. Esto lo hace útil para almacenar datos que cambian a veces.

- Usted puede encontrar EEPROM En dispositivos como microcontroladores. Almacena datos pequeños, como configuraciones, que requieren actualizaciones. Aunque es más lenta que otros tipos de memoria, EEPROM Es confiable para guardar datos importantes.

- Fun Fact: EEPROM Puede soportar miles de ciclos de escritura y borrado, lo que le confiere una larga vida útil.

**El ESP32 dispone 2 tipos de memorias:**
- Memorias internas (Incluidas en el Soc)
- Memorias externas (Se pueden añadir para expandir la capacidad del sistema)

### MEMORIA de ESP32

- **Memoria ROM** 448 KiB (de solo escritura), aqui se almacenan los codigos que gestionan el arranque, la comunicacion del wi-fi y el bluetooh

- **Memoria SRAM interna** 520 KiB, esta memoria es utilizada por el procesador para almcaenar tanto datos como instrucciones y tiene la ventaja que para el procesador es mucho mas facil acceder a esta que al SRAM externa, lo que aumenta la velocidad de proocesamiento

- **Memoria RTC SRAM** 16KiB, utilizados por el co-procsador para realizar operaciones de ultra bajo consumo

- **Memoria Efuse** 256 bits utilizados por el propio sistema y los 768 bits restantes reservados para otras aplicaciones

- **Memoria Flash empotrada (Embedded flash)**, memoria variable destinada al almacenamiento del codigo de nuestra aplicacion y la cantidad de esta memoria varia dependiendo del chip que estemos usando

Si la memoria no fuese suficiente para nuestro proyecto o dispositivo, tenemos la posibilidad de agregar **hasta 16 megabytes de memoria FLASH esterna** y **hasta 8 megabytes de memoria SRAM externa** de esta manera podemos desarrollar aplicaciones todavia mas complejas


## **Pines de la placa**



- La placa cuenta con 25 pines digitales: para conectar leds, botones, reles y un sinfin de perifericos

- La placa cuenta con un conversor analogico-digital ( in ADC out) de 12 bits y 18 canales, es decir, podemos efectuar lecturas de hasta 18 sensores analogicos

- 2 Conversores digital-analogico DAC ( in DAC out), esto permite generar 2 señales de voltaje analogicas puras, que pueden ser utilizadas para controlar un circuito analogico, manipular la intensidad de un led, o incluso agregar un pequeño amplificador y un altavoz a nuestro proyecto para reproducir musica

- Dispone de 9 Pines con sensores tactiles capacitivos que podemos usar a modo de botones, estos pines son capaces de detectar las pequeñas variaciones producidas al acercar un dedo al pin, de esta forma es posible crear todo tipo de controles como botones o baras de deslizamiento sin necesidad de utilizar componentes mecanicos

- Dispone de un controlador SD/SDIO/MMC, esto nos permite interactuar con tarjetas SD, y MMC directamente

- Cuanta con 3 Puertos UART (Comunicacion serial), estos permiten establecer comunicaciones asincronas entre dos dispositivos, utilizando solamente dos pines (TX y RX)

- 2 Interfaces de comunicacion I2C, que admiten los modos de operacion "Master" y "Slave", que pueden funcionar de tres formas diferentes, modo estandar(100 kilowatts/s), modo rapido(400 kilowatts/S), y el direccionamiento de 7 a 10 bits 

- 4 Puertos SPI, pueden alcanzar altas velocidades de hasta 80 megahercios con un buffer de 64 bytes para transmision y recepcion 

- 2 Puertos I2S, para transmision de datos de audio

La placa nos permite utilizar hasta 16 pines como salida PWM

### SEGURIDAD

- El ESP32 cuenta con aceleradores de algoritmos orientados a la encriptacion y soporta los siguientes tipos de cifrado (SHA, RSA, AES, RNG), esto es bastante importante para ciertos proyectos en donde debamos mantener a salvo la informacion


### CONECTIVIDAD INALAMBRICA
- Dispone de Wi-Fi integrado de alto rendimiento, compatible con 802.11 b/g/h. en la banda de los 2.4 GHz alcanzando velocidades de hasta 150 Mbits/s

- Comunicacion Bluetooh compatible con Bluetoohv4.2 y Bluetooh Low Energy (BLE)

### SENSOR DE EFECTO HALL

- Dispone de un sensor de campos magneticos para multiples aplicaciones

### **¿Para qué se usa?**

Algunos ejemplos de proyectos:

- Domótica (encender luces desde el celular).
- Estaciones meteorológicas.
- Robots y vehículos autónomos.
- Sistemas de monitoreo remoto.
- Dispositivos IoT conectados a internet.
- Control de motores y servos.

**Ejemplo sencillo**

Un ESP32 puede leer la temperatura de un sensor y enviar los datos por Wi-Fi a una página web o una aplicación móvil.

**Diferencia entre ESP32 y Arduino Uno**
|Característica	|ESP32	Arduino Uno|
|:-:|:-:|
|Wi-Fi|	Sí|	No|
|Bluetooth|	Sí|	No|
|Velocidad|	Hasta 240 MHz|	16 MHz|
|Memoria|	Mucho mayor|	Menor|
|Voltaje lógico|	3.3 V|	5 V|

Por eso, para proyectos conectados a Internet o aplicaciones más avanzadas, el ESP32 suele ser una opción más potente que un Arduino Uno.
