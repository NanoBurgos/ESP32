#### **Siempre que se desarrolle un proyecto con ESP32 es muy importante tener en cuenta:**

## **EXISTEN PINES CON CARACTERISTICAS ESPECIFICAS QUE LOS HACEN ADECUADOS O NO PARA UN PROYECTO**

Esto es debido al comportamiento que tiene cada pin en su arranque

Cuando nuestra placa es alimentada y se inicia, una serie de procesamientos internos del chip, provocan que ciertos pines cambien su estado a High(Alto) o emitn señales que pueden provocar un efecto no deseado en nuestros proyectos

- Esto significa si que tenemos salidas conectadas a estos GPIO, podemos obtener resultados inesperados cuando el ESP32 se reinicia o arranca

### **ESP32 DEVKIT V1**

Nos centraremos en los pines de esta placa, y los clasificaremos en tres grupos segun su uso como entradas o como salidas




- **El GPI01**, no deberiamos usarlo como entrada, ya que se trata del pin TX del puerto UART de depuracion de la placa, 
  - y en el arranque o en el reinicio o cuando hacemos uso del puerto serie, este emitira datos de depuracion; 
  - Este mismo GPIO si podriamos usarlo como salida, pero teniendo en cuenta que del mismo modo, durante el arranque o reinicio de la placa cambiara su estado a High(Alto)

- **El GPI02**, es el que va conectado al led integrado de la placa, pero podriamos usarlo tanto como entrada como para salida por que no realiza ningun comportamiento extraño

- **El GPI03**, al igual que el GPIO1 es usado por el puerto UART para la depuracion, por lo que no deberiamos usarlo para salida, aunque si podriamos usarlo como entrada teniendo en cuenta su estado High(Alto) durante el arranque o reinicio

- **El GPI04** y **El GPI05** podemos usarlos sin problemas como entradas o como salidas

- **El GPI012** podemos usarlo tambien como entrada o como salida pero debemos tener en cuenta que durante el arranque no debemos ponerlo en estado High(Alto) por que provocara un fallo en el inicio

- **El GPI013** tambien podemos usarlo sin problemas

- **El GPI014** y **El GPI015** tambien podemos usarlo sin problemas aunque en el inicio emitiran señales PWM

- Los **El GPI16** al - **El GPI033** podemos utilizarlos sin ningun impedimento, tanto como entradas o como salidas

- Los **GPI034, GPI035, GPI036, GPI039** podemos utilizarlos unicamente como entradas, ya que estos pines no disponen de resistencias internas Pull-Up o Pull-Down

### **En resumidas cuentas:**
- Los pines resaltados en verde, son los que podemos usar sin ningun tipo de problema, ya que se comportaran segun lo esperado
- Los pines resaltados en amarillo, podemos usarlos pero debemos tener cuidado por que pueden tener un comportamiento no deseado, principalmente en el arranque
- Los pines resaltados en rojo, no se aconseja usar, ya que estos se pondran en estado High(Alto) o emitiran señales que entorpeceran el funcionamiento de nuestros dispositivos
